```python
img = cv2.imread('My Photo.jpeg')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
```

```python
r, g, b = cv2.split(img_rgb)
```

```python
plt.imshow(b, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')
```

```python
print("Blue Channel Shape:", b.shape)
```

```python
print("Image Shape:", img.shape)
print("Image Type:", type(img))
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
_, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis('off')
```

```python
h, w = gray.shape
center_x, center_y = w // 2, h // 2

crop = img[center_y-100:center_y+100, center_x-100:center_x+100]

plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.title("Cropped 200x200")
plt.axis('off')
```

```python
branded = img.copy()

cv2.putText(branded, "Jana", (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (0, 0, 255), 2)

plt.imshow(cv2.cvtColor(branded, cv2.COLOR_BGR2RGB))
plt.title("Branded Image")
plt.axis('off')
```

```python
half = cv2.resize(img, (0,0), fx=0.5, fy=0.5)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(half, cv2.COLOR_BGR2RGB))
plt.title("Half Size")
plt.axis('off')
```

```python
nearest = cv2.resize(img, (300,300), interpolation=cv2.INTER_NEAREST)
bilinear = cv2.resize(img, (300,300), interpolation=cv2.INTER_LINEAR)
bicubic = cv2.resize(img, (300,300), interpolation=cv2.INTER_CUBIC)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(nearest, cv2.COLOR_BGR2RGB))
plt.title("Nearest Neighbor")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(bilinear, cv2.COLOR_BGR2RGB))
plt.title("Bilinear")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(bicubic, cv2.COLOR_BGR2RGB))
plt.title("Bicubic")
plt.axis('off')
```
